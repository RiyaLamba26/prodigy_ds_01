```python
# there are so many countries that it is not looking presentable so we can either use top 10 or bottom 10 for better visualization or convert countires into few groups .
```

```python
# let us go ahead with top 10 countries in various age groups
```

```python
top_10=df1.head(10)
top_10 # by total 
top_65=df1.sort_values(by="65+",ascending=False).head(10) # 65+ column
top_25=df1.sort_values(by="25-64 years",ascending=False).head(10) # 25-64 column
top_15=df1.sort_values(by="15-24 years",ascending=False).head(10) # 15-24column
top_5=df1.sort_values(by="5-14 years",ascending=False).head(10) # 5-14column
top_4=df1.sort_values(by="0-4 years",ascending=False).head(10) # 0-4column
```

```python
plt.figure(figsize=(15,20))
plt.subplot(6,1,1)
sns.barplot(x="Country",y="Total",data=top_10,palette="muted")
plt.title("TOP 10 COUNTRIES")
plt.subplot(6,1,2)
sns.barplot(x="Country",y="65+",data=top_65,palette="muted")
plt.title("Top 10 COUNTRIES HAVING 65+ POPULATION")
plt.subplot(6,1,3)
sns.barplot(x="Country",y="25-64 years",data=top_25,palette="muted")
plt.title("Top 10 COUNTRIES HAVING 25-64 POPULATION")
plt.subplot(6,1,4)
sns.barplot(x="Country",y="15-24 years",data=top_15,palette="muted")
plt.title("Top 10 COUNTRIES HAVING 15-24 POPULATION")
plt.subplot(6,1,5)
sns.barplot(x="Country",y="5-14 years",data=top_5,palette="muted")
plt.title("Top 10 COUNTRIES HAVING 5-14 POPULATION")
plt.subplot(6,1,6)
sns.barplot(x="Country",y="0-4 years",data=top_4,palette="muted")
plt.title("Top 10 COUNTRIES HAVING 0-4 POPULATION")
plt.xticks(rotation=45);
```

```python

bot_10=df1.sort_values(by="Total",ascending=False).tail(10) # total column
bot_65=df1.sort_values(by="65+",ascending=False).tail(10) # 65+ column
bot_25=df1.sort_values(by="25-64 years",ascending=False).tail(10) # 25-64 column
bot_15=df1.sort_values(by="15-24 years",ascending=False).tail(10) # 15-24column
bot_5=df1.sort_values(by="5-14 years",ascending=False).tail(10) # 5-14column
bot_4=df1.sort_values(by="0-4 years",ascending=False).tail(10) # 0-4column
```
